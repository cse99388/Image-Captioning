```python
import pandas as pd
import stanza
import string
import re
```

```python
df = pd.read_excel("telugu.xlsx")
df = df.iloc[1:,1:]
```

```python
def clean_word(t):
    a = t.split(" ")
    b = []
    for i in a :
        b.append(i.replace("\u200c",""))
    return " ".join(b)
```

```python
stop_words = ["అందరూ","అడగండి","అడ్డంగా","అనుమతించు","అయితే","ఉన్నారు", "ఎప్పుడు", "ఏదైనా", "ఏమైనప్పటికి", "కాదు", "కూడా", "గురించి", "చేయగలిగింది", "తర్వాత", "దాదాపు", "నిజంగా", "ప్రకారం", "మధ్య", "మరొక", "మాత్రమే", "వద్ద", "వెంట", "వ్యతిరేకంగా","అందుబాటులో","అడగడం","అనుగుణంగా","అనుమతిస్తుంది","ఇప్పటికే","ఎక్కడైనా","ఎవరైనా","ఏ","ఏమైనప్పటికి","ఒక","కనిపిస్తాయి","కాదు","గా","చుట్టూ","తగిన","తర్వాత","దూరంగా","పై","మధ్య","మరియు","మళ్ళీ","మెచ్చుకో","వద్ద","వేరుగా","సంబంధం"]
def remove_stop_words(t):
    a = t.split(" ")
    b = []
    for i in a :
        if i not in stop_words:
            b.append(i)
    return " ".join(b)
```

```python
def remove_punctuation(text):
    transulator = str.maketrans('','',string.punctuation)
    return text.translate(transulator)
```

```python
def remove_numbers(t):
    t = re.sub(r'\d+','',t)
    return t
```

```python
df.iloc[:,1]
```

```python
te_pos_tagger = stanza.Pipeline('te', processors='tokenize,pos,lemma')
```

```python
d = []
for i in range(len(df)):
    t = clean_word(df.iloc[i,1])
    t = remove_stop_words(t)
    t = remove_punctuation(t)
    t = remove_numbers(t)
    doc = te_pos_tagger(t)
    tl = ""
    tn = []
    for sentence in doc.sentences:
        for word in sentence.words:
            tl = tl + " " + word.lemma
            tn.append(word.text + " - " + word.pos )
    d.append([df.iloc[i,0][1:],tl[1:],tn])
    print(i,end = "\r")
```

```python
pd.DataFrame(d,columns = ["image","clean_text","pos_tag_text"]).to_excel("telugu_clean_pos.xlsx")
```
